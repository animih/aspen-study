```python
%load_ext autoreload
%autoreload 2

import sys

sys.path.append('../nonlin')
sys.path.append('../partion')
sys.path.append('../1D-difusion')

from aspen import aspen
from newton import newton

import partion
from partion import metrics1, metrics2, m1, m2, m3
import wrapper

from one_phase import one_phase
import numpy as np
import matplotlib.pyplot as plt
```

```python
def model(u):
    return np.array([u**2+1, 2*max(u, 0)])

class Diffusion():
    def __init__(self, Nx, Amp=0, Period=0, Scale=1, model=None):
        self.val = np.zeros((Nx+1, 1))
        for i in range(Nx+1):
            self.val[i] = (
                Amp * np.sin(Period * 2*np.pi * (i+1) / (Nx+1)) + 1) * Scale
        self.model = model
```

```python
plt.style.use('dark_background')
# inital condition
Nx = 1000
Nt = 100

D = Diffusion(Nx, 0.9, 1, 0.5e-1, model=model)
bd1 = 0
bd2 = 1

x0 = np.arange(Nx)/(Nx-1)
x0 = np.sinh(20*(x0-0.5))
x0 /= np.max(x0)
x0 = (x0+1)/2
x0 = np.reshape(x0, (-1, 1))
plt.xlabel('x0')
plt.ylabel('u')
plt.plot(x0)

setInitial = lambda x: x#.setInitial(0.2, 1, 1) 
setSources = lambda x: x#.setSources([0.9], [-17.9])
```

```python
param = {
    'Nx': Nx, 
    'Nt': Nt,
}
nl_solver = newton(1e-2, kmax = 14, crit_abs = 1e-3)
solver_cl = one_phase(param, D, nl_solver)
X_cl, message, t_new = wrapper.calc(solver_cl, x0, bd1, bd2)
```

```python
# we treet the value in time at particular coord as a variable
# let's visualize the covariance matrix of these varibles

cov_m = np.cov(X_cl)
cs = plt.matshow(cov_m)
cbar = plt.colorbar(cs)
plt.show()

cs = plt.matshow(np.abs(cov_m))
cbar = plt.colorbar(cs)
plt.show()
```

```python
# hm... no we need to subdivide it into blocks
# we already have spectral bisection implemented why not use it?
borders = []
to_explore = [(0, Nx)]
Nd = 4
for i in range(Nd-1):
    bd = to_explore.pop(0)
    part = partion.spec_bis((cov_m[bd[0]:bd[1], bd[0]:bd[1]]))
    split = np.argmax(np.abs(part))
    borders.append(bd[0]+split)
    to_explore.append((bd[0], bd[0]+split))
    to_explore.append((bd[0]+split, bd[1]))
    
borders = np.sort(borders)
borders = np.concatenate(([0], borders, [Nx]))
```

```python
print(borders)
```

```python
Nd = 4
param = {
    'Nx': Nx, 
    'Nt': Nt
}
nl_solver = aspen(Nd, borders, 1e-2, crit_abs = 1e-3)
solver_as = one_phase(param, D, nl_solver)
X, message, t_1 = wrapper.calc(solver_as, x0, bd1, bd2)
```

```python
wrapper.compare([solver_cl, solver_as], ['newton', 'aspen'])
wrapper.show_res(solver_as)
wrapper.compare([solver_cl, solver_as], ['newton', 'aspen'])
```
