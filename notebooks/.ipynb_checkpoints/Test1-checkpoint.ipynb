```python
%load_ext autoreload
%autoreload 2

import sys

sys.path.append('../nonlin')
sys.path.append('../partion')
sys.path.append('../1D-difusion')

from aspen import aspen
from newton import newton

import partion
from partion import metrics1, metrics2, m1, m2, m3
import wrapper

from one_phase import one_phase
import numpy as np
import matplotlib.pyplot as plt
```

### Начальные условия и модель

```python
def model(u):
    return np.array([u**2+1, 2*max(u, 0)])

class Diffusion():
    def __init__(self, Nx, Amp=0, Period=0, Scale=1, model=None):
        self.val = np.zeros((Nx+1, 1))
        for i in range(Nx+1):
            self.val[i] = (
                Amp * np.sin(Period * 2*np.pi * (i+1) / (Nx+1)) + 1) * Scale
        self.model = model
```

```python
plt.style.use('dark_background')
# inital condition
Nx = 1000
Nt = 100
D = Diffusion(Nx, 0.9, 1, 0.5e-1, model=model)
bd1 = 0
bd2 = 1
x0 = np.arange(Nx)/(Nx-1)
x0 = np.sinh(20*(x0-0.5))
x0 /= np.max(x0)
x0 = (x0+1)/2
x0 = np.reshape(x0, (-1, 1))
plt.xlabel('x0')
plt.ylabel('u')
plt.plot(x0)
```

Такое начальное условие выбрано для динамики, позднее на графиках будет видно как решение меняется по времени

```python
test = wrapper.test_decorator(wrapper.test)
```

```python
param = {
    'Nx': Nx, 
    'Nt': Nt,
}
nl_solver = newton(1e-2, kmax = 14, crit_abs = 1e-3)
solver_cl = one_phase(param, D, nl_solver)
solver_cl.x0 = np.copy(x0)
solver_cl.setBoundary(bd1, bd2)
X_cl, message, t_new = test(solver_cl, x0, sample_size = 5)
```

```python
Nd = 4
param = {
    'Nx': Nx, 
    'Nt': Nt
}
part = partion.partion_equally(Nx, Nd)
nl_solver = aspen(Nd, part, 1e-2, crit_abs = 1e-3)
solver_as = one_phase(param, D, nl_solver)
solver_as.x0 = np.copy(x0)
solver_as.setBoundary(bd1, bd2)
X, message, t_eq = test(solver_as, x0, sample_size = 5)
```

### Графики решения

```python
wrapper.show_res(solver_cl)
wrapper.show_res(solver_as)
x = np.linspace(0, 1, Nx)

plt.title('Diff')
plt.xlabel('x')
plt.ylabel('t')
t = solver_as.t
x_grid, t_grid = np.meshgrid(x, t)
cs = plt.contourf(x_grid, t_grid, (np.abs(X-X_cl)).T, cmap='RdBu_r')
cbar = plt.colorbar(cs)
plt.show()
```

Как видим а) присутсвует динамика по времени, имеющая локальный характер, б) решения двух алгоритмов отличаются в пределах погрешности

### Две формулировки функции на минимизацию:

$$
	vol(\partial S_i) = \sum_{v \in V_i} \sum_{u \in \Gamma(v), u \notin V_i} {w(v, u)}
$$
Первый способ:
$$
\sum \frac{vol(\partial S_i)}{| V_i|} \rightarrow \min
$$
Второй способ:
$$
\sum \frac{vol(\partial S_i)}{vol(V_i)} \rightarrow \min
$$
где $vol(V_i) = \sum_{u, v \in V_i} w(u, v)$

##### Метрика 1
Откровенно неверная, но почему-то хорошо работающая в конкретно нашей задаче
$$
w(v, u) =  (|\frac{\partial R(u)}{\partial v}| + |\frac{\partial R(v)}{\partial u}|)^{-1}
$$
Будем рассмтривать только первую формировку пробелмы для неё (вторая формулировка была была проверена но [данные удалены] из ноутбука, т.к. там не было выгоды)

##### Метод 3
$$
w(v, u) = С_u | \frac{\partial R(v)}{\partial u} | + С_v | \frac{\partial R(u)}{\partial v} |
$$

где $C_u = |\frac{\partial^2 R(u)}{ \partial u^2}|\cdot
|\frac{\partial R(u)}{ \partial u} |^{-2}$

```python
# Matrix of Weights
M = 2
A = np.zeros((Nx, Nx))
du = 1e-2
for i in range(M):
    step = Nt//M
    Jf = partion.precompute_Jf(solver_cl, X_cl[:, i*step], Nx)
    Jf_n = partion.precompute_Jf(solver_cl, X_cl[:, i*step]+du, Nx)
    func = lambda i, j: m3(solver_cl, Jf, Jf_n, i, j)
    B = partion.adj_matrix(func, Nx)
    B /= np.max(B)
    A += B
A /= M
plt.matshow(A[650:750, 650:750])
plt.show()
plt.matshow(A[:200, :200])
plt.show()
```

```python
Nd = 4
borders = partion.domain_builder1(A, Nd, inv = False)
print(borders)

print('--comparison--')
print('eq : {}'.format(partion.metrics1(
    partion.partion_equally(Nx, Nd), A)))
print('spec : {}'.format(partion.metrics1(
    borders, A)))

results['p1_m3_rc'] = np.copy(borders)
```

```python
import csv
gather_data = False

Nd = 4
if gather_data:
    N = 5
    param = {
        'Nx': Nx, 
        'Nt': Nt
    }

    time = np.zeros(N)
    borders = np.zeros((Nd-1, N), dtype = 'int')
    num_of_alghs = 0

    for k in range(N):
        part = partion.generate_random(Nx, Nd)
        borders[:, k] = part[1:-1]
        nl_solver = aspen(Nd, part, 1e-2, crit_abs = 1e-3)
        solver_ts = one_phase(param, D, nl_solver)
        solver_ts.setBoundary(bd1, bd2)
        solver_ts.x0 = np.copy(x0)
        _, _, _, t = wrapper.test(solver_ts, x0, sample_size = 5)
        if(message != 'OK'):
            continue
        time[k] = t
        num_of_alghs += 1
        if k % 10 == 0:
            print(k)      
    # save data
    name = './data/Nx={},Nt={},Nd={}.csv'.format(Nx, Nt, Nd)
    filt = time > 0
    with open(name, 'a') as csvfile:
        writer = csv.writer(csvfile, delimiter=',')
        writer.writerows(borders[:, filt])
        writer.writerow(time[filt])
else:
    name = './data/Nx={},Nt={},Nd={}.csv'.format(Nx, Nt, Nd)
    with open(name, 'r') as csvfile:
        reader = csv.reader(csvfile)
        data = []
        for row in reader:
            data.append(row)        
        borders = np.array(data[0:Nd-1], dtype='int')
        time = np.array(data[Nd-1], dtype= 'float')
        num_of_alghs = np.sum(time > 0)
```

```python
print('Numner og alghorithms that converged:', num_of_alghs)
print('mean time :', np.sum(time)/num_of_alghs)
sq_cov = np.sqrt(np.sum(time**2)/num_of_alghs-(np.sum(time)/num_of_alghs)**2)
print('sqrt cov :', np.sqrt(np.sum(time**2)/num_of_alghs-(np.sum(time)/num_of_alghs)**2))
```

```python
from scipy.stats import gaussian_kde
from scipy.integrate import quad

width = 4
win = np.arange(5, 155, width)
y = []
for x in win:
    mask = ((time < x+width/2) & (time > (x - width/2)))
    y.append(np.sum(mask)/ num_of_alghs/5)
    
plt.bar(win, y, width=width*0.5, color = 'r')
bw = 1/(1.06*sq_cov)*num_of_alghs**(1/5)
filt = time > 0
kernel = gaussian_kde(time[filt], bw_method=bw)

plt.plot(np.linspace(40, 120, 100), kernel(np.linspace(40, 120, 100)))
plt.xlim((40, 120))
plt.axvline(t_new, label='newton', linestyle ='--', color='y')
plt.axvline(t_eq, label='eq', linestyle ='--', color='c')
plt.axvline(results['t_m1'], label='m1', linestyle ='--', color ='r')
plt.axvline(results['t_m2'], label='m2', linestyle ='--', color ='g')
plt.axvline(results['t_m3'], label='m3', linestyle ='--', color ='w')

plt.legend()
plt.show()
```

```python
# we will estimate the efficency by function of distibution
p_eq, err_eq = quad(kernel, 0, t_eq)
print('equally partion : {}'.format(p_eq))
p_m1, err_m1 = quad(kernel, 0, results['t_m1'])
print('m1 partion : {}'.format(p_m1))
p_m2, err_m2 = quad(kernel, 0, results['t_m2'])
print('m2 partion : {}'.format(p_m2))
p_m3, err_m3 = quad(kernel, 0, results['t_m3'])
print('m3 partion : {}'.format(p_m3))
```

```python
filt = time > 0
m1_list = []
m2_list = []
m3_list = []

# Matrix of Weights
# Matrix of Weights
M = 5
A1 = np.zeros((Nx, Nx))
A3 = np.zeros((Nx, Nx))
du = 1e-2
for i in range(M):
    step = Nt//M
    Jf = partion.precompute_Jf(solver_cl, X_cl[:, i*step], Nx)
    Jf_n = partion.precompute_Jf(solver_cl, X_cl[:, i*step]+du, Nx)
    func = lambda i, j: m1(solver_cl, Jf, i, j)
    B = partion.adj_matrix(func, Nx)
    B /= np.max(B)
    A1 += B
    func = lambda k, m: m3(solver_cl, Jf, Jf_n, k, m)
    B = partion.adj_matrix(func, Nx)
    #B /= np.max(B)
    A3 += B
plt.show()
A1 /= M
A3 /= M
func = lambda i, j: m2(solver_cl, X_cl, i, j)
A2 = partion.adj_matrix(func, Nx)

func1 = lambda borders: metrics1(borders, A1)
func2 = lambda borders: metrics1(borders, A2)
func3 = lambda borders: metrics1(borders, A3)

for domain in borders[:, filt].T:
    part = np.concatenate(([0], domain, [Nx]), axis = 0)
    m1_list.append(func1(part))
    m2_list.append(func2(part))
    m3_list.append(func3(part))

m1_arr = np.array(m1_list)
m2_arr = np.array(m2_list)
m3_arr = np.array(m3_list)

plt.title('m1')
plt.scatter(m1_arr, time[filt])
#plt.legend()
plt.xscale('log')
plt.show()
plt.title('m2')
#plt.legend()
plt.scatter(m2_arr, time[filt])
plt.xscale('log')
plt.show()
plt.title('m3')
#plt.legend()
plt.scatter(m3_arr, time[filt])
plt.xscale('log')
plt.show()
```

```python
print('--v---corr---v--')
cov_m1 = np.cov(np.stack((time[filt], m1_arr) , axis = 0))
print('m1 :', cov_m1[0, 1]/np.sqrt(cov_m1[1, 1])/np.sqrt(cov_m1[0, 0]))

cov_m2 = np.cov(np.stack((time[filt], m2_arr) , axis = 0))
print('m2 :', cov_m2[0, 1]/np.sqrt(cov_m2[1, 1])/np.sqrt(cov_m2[0, 0]))

cov_m3 = np.cov(np.stack((time[filt], m3_arr) , axis = 0))
print('m3 :', cov_m3[0, 1]/np.sqrt(cov_m3[1, 1])/np.sqrt(cov_m3[0, 0]))
```

### Зависимость числа иттераций/времени работы от шага по времени

```python
plt.style.use('dark_background')
# inital condition
Nx = 1000
Nt = 100
D = Diffusion(Nx, 0.9, 1, 0.5e-1, model=model)
bd1 = 1
bd2 = 0
x = np.linspace(0, 1, Nx)
x0 = np.exp(-x/2)*np.cos(3*np.pi/2*x)**2
x0 = np.reshape(x0, (-1, 1))
plt.xlabel('x0')
plt.ylabel('u')
plt.plot(x0)
plt.show()
```

```python
newton_time = np.zeros(10)
newton_iters = np.zeros(10)

for i in range(10):
    Nt = 50*(1+i)
    param = {
        'Nx': Nx, 
        'Nt': Nt,
    }
    nl_solver = newton(1e-2, kmax = 14, crit_abs = 1e-3)
    solver_cl = one_phase(param, D, nl_solver)
    solver_cl.x0 = np.copy(x0)
    solver_cl.setBoundary(bd1, bd2)
    _, X_cl, message, t_new = wrapper.test(solver_cl, x0, sample_size = 1)
    
    newton_time[i] = t_new
    newton_iters[i] = np.sum(solver_cl.timelog.kn)/Nt
    if message != 'OK':
        print(i, 'not converged')
```

```python
# Matrix of Weights
M = 4
A1 = np.zeros((Nx, Nx))
for k in range(M):
    step = Nt//M
    Jf = partion.precompute_Jf(solver_cl, X_cl[:, k*step], Nx)
    func1 = lambda i, j: m1(solver_cl, Jf, i, j)
    B = partion.adj_matrix(func1, Nx)
    B /= np.max(B)
    A1 += B
A1 /= M
```

```python
borders = partion.domain_builder1(A1, 6, inv = False)
print(borders)
```

```python
aspen_time = np.zeros(10)
aspen_iters = np.zeros(10)
local_iters = np.zeros((10, 6))

for i in range(10):
    Nt = 50*(1+i)
    param = {
        'Nx': Nx, 
        'Nt': Nt,
    }
    nl_solver = aspen(6, borders, 1e-2, max_lc = 14, crit_abs = 1e-3)
    solver_as = one_phase(param, D, nl_solver)
    solver_as.x0 = np.copy(x0)
    solver_as.setBoundary(bd1, bd2)
    _, _, message, t = wrapper.test(solver_as, x0, sample_size = 1)
    
    aspen_time[i] = t
    aspen_iters[i] = np.sum(solver_as.timelog.aspen_iters)/Nt
    local_iters[i, :] = solver_as.timelog.domain_iters
    if message != 'OK':
        print(i, 'not converged')
```

```python
plt.title('time comparision')
plt.xlabel('$\Delta t$')
dt = 1/(50*np.arange(1, 11))
plt.plot(dt, newton_time, label = 'Newton')
plt.plot(dt, aspen_time, label = 'ASPEN')
plt.legend()
plt.show()
```

```python
plt.title('iters compatison')
plt.xlabel('$\Delta t$')
dt = 1/(50*np.arange(1, 11))
plt.plot(dt, newton_iters*np.arange(1, 11), label = 'newton')
plt.plot(dt, aspen_iters*np.arange(1, 11), label = 'ASPEN')
plt.legend()
plt.show()
```
